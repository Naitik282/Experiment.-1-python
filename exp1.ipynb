Introduction to IDLE
IDLE stands for Integrated Development and Learning Environment.
It is the default Python editor that comes installed with Python.
You can open IDLE by searching IDLE in the Start Menu.

Interactive Mode in IDLE
Opens automatically when IDLE starts.
Shows the prompt symbol >>>.
Executes one line of code at a time.
Output appears immediately after pressing Enter.
Mainly used for quick testing and simple calculations.

Scripting Mode in IDLE
Used to write and run full programs.
Open by clicking File → New File.
Write multiple lines of code.
Save the file with .py extension.
Run the program using Run → Run Module.

Difference Between Interactive and Scripting Mode
Interactive Mode	Scripting Mode
Executes one line at a time	Executes the entire program
Shows >>> prompt	No prompt symbol
No need to save file	Must save file before running
Used for quick testing	Used for full program development

8. Write a program to find area of triangle when length of sides are given.

```python
# Import math module
import math

# Input sides of the triangle
a = float(input("Enter side a: "))
b = float(input("Enter side b: "))
c = float(input("Enter side c: "))

# Check for valid triangle using the triangle inequality theorem
if (a + b > c) and (a + c > b) and (b + c > a):
    # Calculate semi-perimeter
    s = (a + b + c) / 2

    # Calculate area using Heron's formula
    # The term inside sqrt should be non-negative for a valid triangle
    area = math.sqrt(s * (s - a) * (s - b) * (s - c))

    # Print the result
    print("Area of the triangle is:", area)
else:
    print("Error: The given side lengths do not form a valid triangle.")
    print("Please ensure that the sum of the lengths of any two sides is greater than the length of the third side.")
```

Write a program to find sum of first n natural numbers.

```python
# Input the value of n
n = int(input("Enter a positive integer n: "))

# Calculate sum using formula: sum = n*(n+1)/2
sum_natural = n * (n + 1) // 2

# Print the result
print("Sum of first", n, "natural numbers is:", sum_natural)

```

Write a program to find left shift and right shift values of a given number.

```python
# Input the number and shift value
num = int(input("Enter a number: "))
shift = int(input("Enter number of bits to shift: "))

# Left shift (<<)
left_shift = num << shift

# Right shift (>>)
right_shift = num >> shift

# Print results
print("Left shift of", num, "by", shift, "bits is:", left_shift)
print("Right shift of", num, "by", shift, "bits is:", right_shift)
```

Using membership operator find whether a given number is in sequence (10,20,56,78,89)

```python
# Define the sequence
sequence = (10, 20, 56, 78, 89)

# Input a number
num = int(input("Enter a number to check: "))

# Check membership
if num in sequence:
    print(num, "is present in the sequence.")
else:
    print(num, "is NOT present in the sequence.")


```
